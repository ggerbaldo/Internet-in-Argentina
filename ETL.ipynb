```python
# Combinamos dataframes
combinado_penetracion = pd.merge(penetracion_hab, penetracion_hog, on=('Año', 'Trimestre', 'Provincia'))
combinado_penetracion.info()
```

```python
# En la revisión del df tecnologia_provincia, observamos datos NaN en el campo 'Año'y 'Trimestre' con asteríscos (*).
tecnologia_provincia['Año'].unique() # Se revisan valores
tecnologia_provincia.dropna(inplace=True) # Eliminamos nulos
tecnologia_provincia.info()
```

```python
# Función para limpiar y convertir los valores de año y trimestre ,no numéricos, a números
def limpiar_df(valor):
    if isinstance(valor, str):
        try:
            return int(valor.split()[0])
        except ValueError:
            return 0  
    return valor

#Seleccionar las columnas de interés
columnas = ['Año', 'Trimestre']

```

```python
#Aplicar la función de limpieza a las columnas de interés
for col in columnas:
   tecnologia_provincia[col] = tecnologia_provincia[col].apply(limpiar_df)
```

```python
# Unimos los df combinado_penetraciion y tecnologia_provincia
combinado_1 = pd.merge(tecnologia_provincia, combinado_penetracion, on=['Año', 'Trimestre', 'Provincia'])
combinado_1.head(5)
```

```python
# Combinamos el df con los tipo de servicios Dial Up y Banda ancha
combinado_2 = pd.merge(combinado_1, Dial_BAf_prov, on=['Año', 'Trimestre', 'Provincia'])
combinado_2.info()
```

```python
# Antes de sumar el df de velocidad x provincia, debemos corregir el año del cuarto trimestre de 2023, que aparece erróneamente como 2024.
velocidad_prov['Año'] = velocidad_prov['Año'].replace(2024, 2023)
```

```python
# Adicionamos los rangos de velocidad
combinado_3 = pd.merge(combinado_2, velocidad_prov, on=['Año', 'Trimestre', 'Provincia'], how='left')
combinado_3.info()
```

```python
velocidad_media_provincia.info()
```

```python
# Sumamos la velocidad media por provincia
df_provincia = pd.merge(combinado_3, velocidad_media_provincia, on=['Año', 'Trimestre', 'Provincia'])
df_provincia.info()
```

```python
# Eliminamos columna de total repetidas y renombramos, junto a campos 'Otros'.
df_provincia.drop('Total_x', axis=1, inplace=True)
df_provincia.drop('Total', axis=1, inplace=True)
df_provincia.rename(columns={'Total_y': 'Total'}, inplace=True)
df_provincia.rename(columns={'Otros': 'otros_tec'}, inplace=True)
df_provincia.rename(columns={'OTROS': 'otros_vel'}, inplace=True)
df_provincia.info()
```
